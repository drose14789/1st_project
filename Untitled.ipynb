```python
# ============================================================
# 1) ÎùºÏù¥Î∏åÎü¨Î¶¨ Î°úÎìú
# ============================================================
import numpy as np  # ÏàòÏπò Í≥ÑÏÇ∞(Î∞∞Ïó¥/ÏÇ¨Ïù∏ÏΩîÏÇ¨Ïù∏ Îì±)
import pandas as pd  # Îç∞Ïù¥ÌÑ∞ Î°úÎìú/Ï†ÑÏ≤òÎ¶¨/ÏßëÍ≥Ñ
import geopandas as gpd  # ÏßÄÎèÑ(Íµ¨ Í≤ΩÍ≥Ñ) Î°úÎìú/Ï°∞Ïù∏
import matplotlib.pyplot as plt  # ÏãúÍ∞ÅÌôî(ÏßÄÎèÑ/Í∑∏ÎûòÌîÑ)

from pytorch_tabnet.tab_model import TabNetClassifier  # TabNet Î∂ÑÎ•ò Î™®Îç∏

from sklearn.metrics import roc_auc_score  # ROC-AUC
from sklearn.metrics import average_precision_score  # PR-AUC(AP)
from sklearn.metrics import log_loss  # LogLoss
from sklearn.metrics import brier_score_loss  # Brier Score
from sklearn.metrics import precision_recall_curve  # PR Ïª§Î∏å(ÏûÑÍ≥ÑÍ∞íÎ≥Ñ precision/recall)
from sklearn.metrics import roc_curve  # ROC Ïª§Î∏å(ÏûÑÍ≥ÑÍ∞íÎ≥Ñ TPR/FPR)
from sklearn.metrics import confusion_matrix  # ÌòºÎèôÌñâÎ†¨
from sklearn.metrics import precision_score  # Precision
from sklearn.metrics import recall_score  # Recall
from sklearn.metrics import f1_score  # F1
from sklearn.metrics import accuracy_score  # Accuracy
from sklearn.calibration import calibration_curve  # Ï∫òÎ¶¨Î∏åÎ†àÏù¥ÏÖò(Ïã†Î¢∞ÎèÑ Í≥°ÏÑ†)

# ===============================
# üìå Í≤ΩÍ≥† Î©îÏãúÏßÄ Ï†úÍ±∞ (ÎåÄÏö©Îüâ CSV Ï≤òÎ¶¨ Ïãú ÍπîÎÅî)
# ===============================
import warnings
warnings.filterwarnings("ignore")

# ===============================
# üìå ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï (Windows)
# ===============================
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
# ============================================================
# 2) Îç∞Ïù¥ÌÑ∞ Î°úÎìú + ÎùºÎ≤® Ï†ïÎ¶¨(0/1Îßå ÏÇ¨Ïö©, 3ÏùÄ Ï†úÏô∏)
# ============================================================
DATA_PATH = r"C:\ai\lecNote\1st_Project\data\ÏÑúÏö∏_ÏùºÎ∞òÏùåÏãùÏ†ê_ÎùºÎ≤®3Ï≤òÎ¶¨.csv"  # Îç∞Ïù¥ÌÑ∞ ÌååÏùº Í≤ΩÎ°ú ÏßÄÏ†ï
df_raw = pd.read_csv(DATA_PATH, low_memory=False)  # CSV Î°úÎìú

if "ÌèêÏóÖ_3ÎÖÑÏù¥ÎÇ¥" not in df_raw.columns:  # ÌÉÄÍπÉ Ïª¨Îüº Ï°¥Ïû¨ ÌôïÏù∏
    raise ValueError("Ïª¨Îüº 'ÌèêÏóÖ_3ÎÖÑÏù¥ÎÇ¥'Í∞Ä ÏóÜÏäµÎãàÎã§.")  # ÏóÜÏúºÎ©¥ Ï¶âÏãú Ï§ëÎã®

df = df_raw.copy()  # ÏõêÎ≥∏ Î≥¥Ï°¥ÏùÑ ÏúÑÌï¥ Î≥µÏÇ¨
df = df[df["ÌèêÏóÖ_3ÎÖÑÏù¥ÎÇ¥"].isin([0, 1])].copy()  # ÎùºÎ≤® 0/1Îßå ÏÇ¨Ïö©(3 Ï†úÏô∏)
df["y"] = df["ÌèêÏóÖ_3ÎÖÑÏù¥ÎÇ¥"].astype(int)  # ÌïôÏäµÏö© ÌÉÄÍπÉ ÏÉùÏÑ±
```

```python
# ============================================================
# 3) Ï∞ΩÏóÖÏãúÏ†ê ÌîºÏ≤ò ÏÑ†ÌÉù(ÎàÑÏàò ÏóÜÎäî Í∏∞Î≥∏Ìòï)
# ============================================================
feature_cols = [  # Ï∞ΩÏóÖÏãúÏ†êÏóê Ïïå Ïàò ÏûàÎäî ÌîºÏ≤òÎßå ÏÇ¨Ïö©
    "Íµ¨",  # ÏßÄÏó≠(Î≤îÏ£ºÌòï)
    "ÏóÖÌÉú_Í∑∏Î£π",  # ÏóÖÌÉú Í∑∏Î£π(Î≤îÏ£ºÌòï)
    "Ï∞ΩÏóÖÏó∞ÎèÑ",  # Ïó∞ÎèÑ(ÏàòÏπòÌòï)
    "Ï∞ΩÏóÖÏõî",  # Ïõî(ÏàòÏπòÌòï/Ï£ºÍ∏∞ÏÑ±)
    "Ï¢åÌëúÏ†ïÎ≥¥x(epsg5174)",  # X Ï¢åÌëú(ÏàòÏπòÌòï)
    "Ï¢åÌëúÏ†ïÎ≥¥y(epsg5174)",  # Y Ï¢åÌëú(ÏàòÏπòÌòï)
]

feature_cols = [c for c in feature_cols if c in df.columns]  # Ïã§Ï†ú Ï°¥Ïû¨ÌïòÎäî Ïª¨ÎüºÎßå ÎÇ®ÍπÄ

need_cols = ["Íµ¨", "ÏóÖÌÉú_Í∑∏Î£π", "Ï∞ΩÏóÖÏó∞ÎèÑ"]  # ÏµúÏÜå ÌïÑÏàò Ïª¨Îüº Ï†ïÏùò
for c in need_cols:  # ÌïÑÏàò Ïª¨Îüº Ï≤¥ÌÅ¨
    if c not in df.columns:  # ÏóÜÏúºÎ©¥
        raise ValueError(f"ÌïÑÏàò Ïª¨Îüº '{c}'Í∞Ä ÏóÜÏäµÎãàÎã§. ÌòÑÏû¨ Ïª¨ÎüºÎ™ÖÏùÑ ÌôïÏù∏Ìï¥ Ï£ºÏÑ∏Ïöî.")  # Ï§ëÎã®

cat_cols = [c for c in ["Íµ¨", "ÏóÖÌÉú_Í∑∏Î£π"] if c in feature_cols]  # Î≤îÏ£ºÌòï Ïª¨Îüº Î™©Î°ù
num_cols = [c for c in feature_cols if c not in cat_cols]  # ÏàòÏπòÌòï Ïª¨Îüº Î™©Î°ù
```

```python
# ============================================================
# 4) Í≤∞Ï∏° Ï≤òÎ¶¨ + (ÏòµÏÖò) Ï∞ΩÏóÖÏõî sin/cos Î≥ÄÌôò
# ============================================================
for c in cat_cols:  # Î≤îÏ£ºÌòï Í≤∞Ï∏° Ï≤òÎ¶¨
    df[c] = df[c].fillna("Unknown").astype(str)  # Í≤∞Ï∏°ÏùÄ UnknownÏúºÎ°ú, Î¨∏ÏûêÏó¥ ÌÜµÏùº

for c in num_cols:  # ÏàòÏπòÌòï Í≤∞Ï∏° Ï≤òÎ¶¨
    df[c] = pd.to_numeric(df[c], errors="coerce")  # Ïà´Ïûê Î≥ÄÌôò(Î∂àÍ∞Ä -> NaN)
    df[c] = df[c].fillna(df[c].median())  # Í≤∞Ï∏°ÏùÄ Ï§ëÏïôÍ∞íÏúºÎ°ú Ï±ÑÏõÄ

if "Ï∞ΩÏóÖÏõî" in df.columns and "Ï∞ΩÏóÖÏõî" in feature_cols:  # Ï∞ΩÏóÖÏõîÏù¥ ÌîºÏ≤òÏóê ÏûàÏúºÎ©¥
    m = df["Ï∞ΩÏóÖÏõî"].clip(1, 12).astype(float)  # Ïõî Î≤îÏúÑÎ•º 1~12Î°ú Ï†úÌïú
    df["Ï∞ΩÏóÖÏõî_sin"] = np.sin(2 * np.pi * m / 12.0)  # Ïõî Ï£ºÍ∏∞ÏÑ± ÏÇ¨Ïù∏
    df["Ï∞ΩÏóÖÏõî_cos"] = np.cos(2 * np.pi * m / 12.0)  # Ïõî Ï£ºÍ∏∞ÏÑ± ÏΩîÏÇ¨Ïù∏
    feature_cols = [c for c in feature_cols if c != "Ï∞ΩÏóÖÏõî"] + ["Ï∞ΩÏóÖÏõî_sin", "Ï∞ΩÏóÖÏõî_cos"]  # Ïõî ÎåÄÏã† sin/cos ÏÇ¨Ïö©
    num_cols = [c for c in feature_cols if c not in cat_cols]  # ÏàòÏπòÌòï Ïª¨Îüº Î™©Î°ù Ïû¨Í≥ÑÏÇ∞
```

```python
# ============================================================
# 5) ÏãúÍ∞Ñ Í∏∞Ï§Ä Split(ÏòàÏãú)
#    - ÎùºÎ≤® Ï†ïÏùòÏÉÅ ÏµúÍ∑º Ïó∞ÎèÑÎäî Î∂ÑÌè¨Í∞Ä Íπ®Ïßà Ïàò ÏûàÏñ¥ <=2022Îßå ÏÇ¨Ïö©(Í∂åÏû•)
# ============================================================
if "Ï∞ΩÏóÖÏó∞ÎèÑ" not in df.columns:  # Ï∞ΩÏóÖÏó∞ÎèÑ ÏóÜÏúºÎ©¥
    raise ValueError("Ïª¨Îüº 'Ï∞ΩÏóÖÏó∞ÎèÑ'Í∞Ä ÏóÜÏäµÎãàÎã§. ÏãúÍ∞Ñ Í∏∞Ï§Ä splitÏùÑ ÏúÑÌï¥ ÌïÑÏöîÌï©ÎãàÎã§.")  # Ï§ëÎã®

df = df[df["Ï∞ΩÏóÖÏó∞ÎèÑ"] <= 2022].copy()  # 2022 Ïù¥ÌïòÎßå ÏÇ¨Ïö©(Í∂åÏû•)

train_df = df[df["Ï∞ΩÏóÖÏó∞ÎèÑ"] <= 2020].copy()  # ÌïôÏäµ: 2020 Ïù¥Ìïò
valid_df = df[df["Ï∞ΩÏóÖÏó∞ÎèÑ"] == 2021].copy()  # Í≤ÄÏ¶ù: 2021
test_df  = df[df["Ï∞ΩÏóÖÏó∞ÎèÑ"] == 2022].copy()  # ÌÖåÏä§Ìä∏: 2022

print("Split rows:", len(train_df), len(valid_df), len(test_df))  # Î∂ÑÌï† ÌÅ¨Í∏∞ Ï∂úÎ†•

test_raw = test_df.copy()  # (Ï§ëÏöî) Ïù∏ÏΩîÎî© Ï†Ñ ÌÖåÏä§Ìä∏ ÏõêÎ≥∏ Î≥¥Í¥Ä(Íµ¨ Îã®ÏúÑ ÌèâÍ∞ÄÏö©)
```

```python
# ============================================================
# 6) Î≤îÏ£ºÌòï Ïù∏ÏΩîÎî©(Í∞ÑÎã® Îß§Ìïë) + TabNetÏö© categorical Ï†ïÎ≥¥ ÏÉùÏÑ±
# ============================================================
enc_maps = {}  # Î≤îÏ£ºÌòï Ïª¨ÎüºÎ≥Ñ mapping Ï†ÄÏû•(Ï†ÑÏ≤¥ ÏòàÏ∏°ÏóêÎèÑ Ïû¨ÏÇ¨Ïö©)

train_enc = train_df.copy()  # Ïù∏ÏΩîÎî©Ïö© ÌïôÏäµ DF
valid_enc = valid_df.copy()  # Ïù∏ÏΩîÎî©Ïö© Í≤ÄÏ¶ù DF
test_enc  = test_df.copy()  # Ïù∏ÏΩîÎî©Ïö© ÌÖåÏä§Ìä∏ DF

for c in cat_cols:  # Î≤îÏ£ºÌòï Ïª¨ÎüºÎ≥Ñ Î∞òÎ≥µ
    cats = pd.Index(train_enc[c].unique())  # trainÏóêÏÑú Î≥∏ Ïπ¥ÌÖåÍ≥†Î¶¨ ÏàòÏßë
    if "Unknown" not in cats:  # UnknownÏù¥ ÏóÜÏúºÎ©¥
        cats = cats.append(pd.Index(["Unknown"]))  # Unknown Ï∂îÍ∞Ä(ÎØ∏Îì±Î°ù Ïπ¥ÌÖåÍ≥†Î¶¨ ÎåÄÎπÑ)
    mapping = {k: i for i, k in enumerate(cats)}  # Ïπ¥ÌÖåÍ≥†Î¶¨ -> Ï†ïÏàò Ïù∏Îç±Ïä§ Îß§Ìïë
    enc_maps[c] = mapping  # Îß§Ìïë Ï†ÄÏû•
    train_enc[c] = train_enc[c].map(lambda x: mapping.get(x, mapping["Unknown"])).astype(int)  # ÌïôÏäµ Ïù∏ÏΩîÎî©
    valid_enc[c] = valid_enc[c].map(lambda x: mapping.get(x, mapping["Unknown"])).astype(int)  # Í≤ÄÏ¶ù Ïù∏ÏΩîÎî©
    test_enc[c]  = test_enc[c].map(lambda x: mapping.get(x, mapping["Unknown"])).astype(int)  # ÌÖåÏä§Ìä∏ Ïù∏ÏΩîÎî©

X_train = train_enc[feature_cols].values  # ÌïôÏäµ ÏûÖÎ†•
y_train = train_enc["y"].values  # ÌïôÏäµ Ï†ïÎãµ

X_valid = valid_enc[feature_cols].values  # Í≤ÄÏ¶ù ÏûÖÎ†•
y_valid = valid_enc["y"].values  # Í≤ÄÏ¶ù Ï†ïÎãµ

X_test = test_enc[feature_cols].values  # ÌÖåÏä§Ìä∏ ÏûÖÎ†•
y_test = test_enc["y"].values  # ÌÖåÏä§Ìä∏ Ï†ïÎãµ

cat_idxs = [feature_cols.index(c) for c in cat_cols]  # TabNetÏóê ÎÑòÍ∏∏ Î≤îÏ£ºÌòï Ïù∏Îç±Ïä§
cat_dims = [len(enc_maps[c]) for c in cat_cols]  # TabNetÏóê ÎÑòÍ∏∏ Î≤îÏ£ºÌòï Ïπ¥ÎîîÎÑêÎ¶¨Ìã∞
```

```python
# ============================================================
# 7) TabNet ÌïôÏäµ
# ============================================================
clf = TabNetClassifier(  # Î™®Îç∏ ÏÉùÏÑ±
    n_d=16,  # decision Ï∞®Ïõê
    n_a=16,  # attention Ï∞®Ïõê
    n_steps=5,  # step Ïàò
    gamma=1.5,  # ÌîºÏ≤ò Ïû¨ÏÇ¨Ïö© ÌóàÏö©ÎèÑ
    lambda_sparse=1e-4,  # Ìù¨ÏÜåÏÑ± Ï†ïÍ∑úÌôî
    cat_idxs=cat_idxs,  # Î≤îÏ£ºÌòï Ïù∏Îç±Ïä§
    cat_dims=cat_dims,  # Î≤îÏ£ºÌòï Ïπ¥ÎîîÎÑêÎ¶¨Ìã∞
    cat_emb_dim=8,  # ÏûÑÎ≤†Îî© Ï∞®Ïõê
    optimizer_params=dict(lr=2e-2),  # ÌïôÏäµÎ•†
    mask_type="sparsemax"  # Ìù¨ÏÜå ÎßàÏä§ÌÅ¨
)

clf.fit(  # ÌïôÏäµ Ïã§Ìñâ
    X_train, y_train,  # ÌïôÏäµ Îç∞Ïù¥ÌÑ∞
    eval_set=[(X_valid, y_valid)],  # Í≤ÄÏ¶ù Îç∞Ïù¥ÌÑ∞
    eval_metric=["auc"],  # ÌïôÏäµ Ï§ë Î™®ÎãàÌÑ∞ÎßÅ ÏßÄÌëú
    max_epochs=200,  # ÏµúÎåÄ epoch
    patience=20,  # early stopping
    batch_size=2048,  # Î∞∞Ïπò
    virtual_batch_size=256,  # ghost batchnorm
    num_workers=0  # ÏõåÏª§ Ïàò
)
```

```python
# ============================================================
# 8) ÌÖåÏä§Ìä∏ ÏÑ±Îä• ÌèâÍ∞Ä(Í∞ÄÍ≤å Îã®ÏúÑ): ROC-AUC / PR-AUC / LogLoss / Brier / Calibration / Threshold / TopK
# ============================================================
p_test = clf.predict_proba(X_test)[:, 1]  # ÌÖåÏä§Ìä∏ ÏòàÏ∏°ÌôïÎ•†(class=1)

y_test = np.asarray(y_test).astype(int)  # Ï†ïÎãµÏùÑ intÎ°ú ÌÜµÏùº
p_test = np.asarray(p_test).astype(float)  # ÌôïÎ•†ÏùÑ floatÎ°ú ÌÜµÏùº

roc = roc_auc_score(y_test, p_test)  # ROC-AUC
pr_auc = average_precision_score(y_test, p_test)  # PR-AUC(AP)
ll = log_loss(y_test, p_test, eps=1e-15)  # LogLoss
brier = brier_score_loss(y_test, p_test)  # Brier Score

print("\n=== Probabilistic metrics (TEST, store-level) ===")  # Ìó§Îçî Ï∂úÎ†•
print(f"ROC-AUC   : {roc:.6f}")  # ROC-AUC Ï∂úÎ†•
print(f"PR-AUC(AP): {pr_auc:.6f}")  # PR-AUC Ï∂úÎ†•
print(f"LogLoss   : {ll:.6f}")  # LogLoss Ï∂úÎ†•
print(f"Brier     : {brier:.6f}")  # Brier Score Ï∂úÎ†•

n_bins = 10  # bin Í∞úÏàò
frac_pos, mean_pred = calibration_curve(y_test, p_test, n_bins=n_bins, strategy="uniform")  # Ïã†Î¢∞ÎèÑ Í≥°ÏÑ† Í∞í

plt.figure(figsize=(6, 6))  # Í∑∏Î¶º ÌÅ¨Í∏∞
plt.plot(mean_pred, frac_pos, marker="o")  # Ï∫òÎ¶¨Î∏åÎ†àÏù¥ÏÖò Í≥°ÏÑ†
plt.plot([0, 1], [0, 1], linestyle="--")  # Ïù¥ÏÉÅÏ†Å ÎåÄÍ∞ÅÏÑ†
plt.xlabel("Mean predicted probability")  # xÏ∂ï
plt.ylabel("Fraction of positives")  # yÏ∂ï
plt.title("Calibration (Reliability Diagram) - TEST")  # Ï†úÎ™©
plt.tight_layout()  # Î†àÏù¥ÏïÑÏõÉ
plt.show()  # Ï∂úÎ†•

bin_edges = np.linspace(0.0, 1.0, n_bins + 1)  # bin Í≤ΩÍ≥Ñ
bin_ids = np.digitize(p_test, bin_edges[1:-1], right=False)  # Í∞Å ÏÉòÌîåÏùò bin id

ece = 0.0  # ECE Ï¥àÍ∏∞Ìôî
mce = 0.0  # MCE Ï¥àÍ∏∞Ìôî
N = len(p_test)  # Ï†ÑÏ≤¥ ÏÉòÌîå Ïàò

for b in range(n_bins):  # bin Î∞òÎ≥µ
    idx = (bin_ids == b)  # bin ÎßàÏä§ÌÅ¨
    nb = idx.sum()  # bin ÌëúÎ≥∏ Ïàò
    if nb == 0:  # ÎπÑÎ©¥
        continue  # Ïä§ÌÇµ
    acc_b = y_test[idx].mean()  # Ïã§Ï†ú ÏñëÏÑ± ÎπÑÏú®
    conf_b = p_test[idx].mean()  # ÌèâÍ∑† ÏòàÏ∏°ÌôïÎ•†
    gap = abs(acc_b - conf_b)  # Ï∞®Ïù¥
    ece += (nb / N) * gap  # ECE ÎàÑÏ†Å
    mce = max(mce, gap)  # MCE Í∞±Ïã†

print("\n=== Calibration summary (TEST, store-level) ===")  # Ìó§Îçî Ï∂úÎ†•
print(f"ECE (bin-based): {ece:.6f}")  # ECE Ï∂úÎ†•
print(f"MCE (bin-based): {mce:.6f}")  # MCE Ï∂úÎ†•

prec, rec, thr = precision_recall_curve(y_test, p_test)  # PR Ïª§Î∏å
prec_t = prec[:-1]  # threshold ÎåÄÏùë precision
rec_t = rec[:-1]  # threshold ÎåÄÏùë recall

f1_t = 2 * (prec_t * rec_t) / (prec_t + rec_t + 1e-12)  # thresholdÎ≥Ñ F1
best_f1_idx = int(np.argmax(f1_t))  # ÏµúÎåÄ F1 Ïù∏Îç±Ïä§
best_f1_thr = float(thr[best_f1_idx])  # Best-F1 threshold
best_f1_val = float(f1_t[best_f1_idx])  # Best-F1 Í∞í

fpr, tpr, thr_roc = roc_curve(y_test, p_test)  # ROC Ïª§Î∏å
youden_j = tpr - fpr  # Youden's J
best_j_idx = int(np.argmax(youden_j))  # ÏµúÎåÄ J Ïù∏Îç±Ïä§
best_j_thr = float(thr_roc[best_j_idx])  # Youden-J threshold

target_recall = 0.80  # Î™©Ìëú recall(ÏõêÌïòÎ©¥ Î≥ÄÍ≤Ω)
eligible = np.where(rec_t >= target_recall)[0]  # Î™©Ìëú recall ÎßåÏ°± Ïù∏Îç±Ïä§
if len(eligible) > 0:  # ÎßåÏ°±ÌïòÎäî thresholdÍ∞Ä ÏûàÏúºÎ©¥
    best_prec_idx = eligible[np.argmax(prec_t[eligible])]  # Í∑∏Ï§ë precision ÏµúÎåÄ
    target_rec_thr = float(thr[best_prec_idx])  # Î™©Ìëú recall threshold
else:  # ÏóÜÏúºÎ©¥
    target_rec_thr = None  # None Ï≤òÎ¶¨

print("\n=== Threshold candidates (TEST, store-level) ===")  # Ìó§Îçî Ï∂úÎ†•
print(f"Default threshold: 0.50")  # Í∏∞Î≥∏ threshold
print(f"Best-F1 threshold: {best_f1_thr:.6f} (F1={best_f1_val:.6f})")  # Best-F1
print(f"Youden-J threshold: {best_j_thr:.6f}")  # Youden-J
print(f"Target recall {target_recall:.2f} threshold: {target_rec_thr}")  # Î™©Ìëú recall threshold

def eval_at_threshold(y_true, p_prob, threshold, name=""):  # threshold ÌèâÍ∞Ä Ìï®Ïàò
    y_pred = (p_prob >= threshold).astype(int)  # thresholdÎ°ú ÏòàÏ∏° ÌÅ¥ÎûòÏä§ ÏÉùÏÑ±
    cm = confusion_matrix(y_true, y_pred)  # ÌòºÎèôÌñâÎ†¨
    pre = precision_score(y_true, y_pred, zero_division=0)  # precision
    re = recall_score(y_true, y_pred, zero_division=0)  # recall
    f1 = f1_score(y_true, y_pred, zero_division=0)  # f1
    acc = accuracy_score(y_true, y_pred)  # accuracy
    print(f"\n--- {name} (threshold={threshold:.6f}) ---")  # Íµ¨Î∂Ñ Ï∂úÎ†•
    print("Confusion Matrix [[TN, FP],[FN, TP]]:")  # ÌòºÎèôÌñâÎ†¨ ÎùºÎ≤®
    print(cm)  # ÌòºÎèôÌñâÎ†¨ Ï∂úÎ†•
    print(f"Precision: {pre:.6f}")  # precision Ï∂úÎ†•
    print(f"Recall   : {re:.6f}")  # recall Ï∂úÎ†•
    print(f"F1       : {f1:.6f}")  # f1 Ï∂úÎ†•
    print(f"Accuracy : {acc:.6f}")  # accuracy Ï∂úÎ†•

eval_at_threshold(y_test, p_test, 0.50, name="Default")  # 0.5
eval_at_threshold(y_test, p_test, best_f1_thr, name="Best-F1")  # Best-F1
eval_at_threshold(y_test, p_test, best_j_thr, name="Youden-J")  # Youden-J
if target_rec_thr is not None:  # Î™©Ìëú recall thresholdÍ∞Ä ÏûàÏúºÎ©¥
    eval_at_threshold(y_test, p_test, target_rec_thr, name=f"Target-Recall({target_recall:.2f})")  # Target Recall

topk_list = [0.01, 0.05, 0.10, 0.20]  # ÏÉÅÏúÑ 1%, 5%, 10%, 20%
order = np.argsort(-p_test)  # ÏòàÏ∏°ÌôïÎ•† ÎÇ¥Î¶ºÏ∞®Ïàú Ïù∏Îç±Ïä§
y_sorted = y_test[order]  # Í∞ôÏùÄ ÏàúÏÑúÎ°ú Ï†ïÎãµ Ï†ïÎ†¨
positives = y_test.sum()  # Ï†ÑÏ≤¥ ÏñëÏÑ±(ÌèêÏóÖ) Ïàò

print("\n=== Recall@TopK (TEST, store-level) ===")  # Ìó§Îçî Ï∂úÎ†•
for frac in topk_list:  # Í∞Å ÎπÑÏú® Î∞òÎ≥µ
    k = int(np.ceil(len(y_test) * frac))  # top-k Í∞úÏàò
    k = max(k, 1)  # ÏµúÏÜå 1Í∞ú Î≥¥Ïû•
    hit = y_sorted[:k].sum()  # top-k Ïïà Ïã§Ï†ú ÌèêÏóÖ Ïàò
    recall_at_k = hit / positives if positives > 0 else 0.0  # Recall@TopK
    print(f"Top {int(frac*100):>2d}% (k={k:>6d})  Recall@TopK: {recall_at_k:.6f}")  # Ï∂úÎ†•
```

```python
# ============================================================
# 9) Íµ¨ Îã®ÏúÑ(ÏßëÍ≥Ñ) ÌèâÍ∞Ä ÏßÄÌëú(TEST): MAE/RMSE + Spearman + Top-K overlap
# ============================================================
if "Íµ¨" not in test_raw.columns:  # test_rawÏóê Íµ¨ Ïª¨ÎüºÏù¥ ÏóÜÏúºÎ©¥
    raise ValueError("test_rawÏóê 'Íµ¨' Ïª¨ÎüºÏù¥ ÏóÜÏäµÎãàÎã§. Íµ¨ Îã®ÏúÑ ÌèâÍ∞ÄÎ•º ÏúÑÌï¥ ÌïÑÏöîÌï©ÎãàÎã§.")  # Ï§ëÎã®

eval_df = test_raw[["Íµ¨"]].copy()  # Íµ¨Îßå Î≥µÏÇ¨
eval_df["y_true"] = y_test  # Ï†ïÎãµ Í≤∞Ìï©
eval_df["p_pred"] = p_test  # ÌôïÎ•† Í≤∞Ìï©

eval_df["Íµ¨"] = eval_df["Íµ¨"].astype(str)  # Î¨∏ÏûêÏó¥ ÌÜµÏùº
eval_df["Íµ¨"] = eval_df["Íµ¨"].apply(lambda s: s if s.endswith("Íµ¨") else s + "Íµ¨")  # 'Íµ¨' ÌëúÍ∏∞ ÌÜµÏùº

gu_actual = eval_df.groupby("Íµ¨")["y_true"].mean().rename("actual_rate")  # Íµ¨Î≥Ñ Ïã§Ï†ú ÌèêÏóÖÎ•†
gu_pred = eval_df.groupby("Íµ¨")["p_pred"].mean().rename("pred_rate")  # Íµ¨Î≥Ñ ÏòàÏ∏° ÌèêÏóÖÎ•†(ÌôïÎ•† ÌèâÍ∑†)
gu_n = eval_df.groupby("Íµ¨")["y_true"].size().rename("n")  # Íµ¨Î≥Ñ ÌëúÎ≥∏ Ïàò

gu_cmp = pd.concat([gu_actual, gu_pred, gu_n], axis=1).reset_index()  # ÎπÑÍµê ÌÖåÏù¥Î∏î

gu_cmp["abs_err"] = (gu_cmp["pred_rate"] - gu_cmp["actual_rate"]).abs()  # Ï†àÎåÄÏò§Ï∞®
gu_cmp["sq_err"] = (gu_cmp["pred_rate"] - gu_cmp["actual_rate"]) ** 2  # Ï†úÍ≥±Ïò§Ï∞®

mae_unweighted = gu_cmp["abs_err"].mean()  # Íµ¨ ÎèôÏùº Í∞ÄÏ§ë MAE
rmse_unweighted = np.sqrt(gu_cmp["sq_err"].mean())  # Íµ¨ ÎèôÏùº Í∞ÄÏ§ë RMSE

w = gu_cmp["n"].astype(float)  # Í∞ÄÏ§ëÏπò(Íµ¨Î≥Ñ ÏÉòÌîå Ïàò)
mae_weighted = (gu_cmp["abs_err"] * w).sum() / w.sum()  # ÏÉòÌîå Ïàò Í∞ÄÏ§ë MAE
rmse_weighted = np.sqrt((gu_cmp["sq_err"] * w).sum() / w.sum())  # ÏÉòÌîå Ïàò Í∞ÄÏ§ë RMSE

spearman_corr = gu_cmp["actual_rate"].rank().corr(gu_cmp["pred_rate"].rank(), method="pearson")  # Spearman(ÏàúÏúÑ ÏÉÅÍ¥Ä)

K = 5  # ÏÉÅÏúÑ KÍ∞ú Íµ¨ ÎπÑÍµê
topk_pred = set(gu_cmp.sort_values("pred_rate", ascending=False).head(K)["Íµ¨"].tolist())  # ÏòàÏ∏° TOP-K Íµ¨
topk_actual = set(gu_cmp.sort_values("actual_rate", ascending=False).head(K)["Íµ¨"].tolist())  # Ïã§Ï†ú TOP-K Íµ¨
hit_k = len(topk_pred.intersection(topk_actual))  # Í≤πÏπú Í∞úÏàò
jaccard_k = hit_k / len(topk_pred.union(topk_actual)) if len(topk_pred.union(topk_actual)) > 0 else 0.0  # Jaccard

print("\n=== District-level metrics (TEST, aggregated by Íµ¨) ===")  # Ìó§Îçî Ï∂úÎ†•
print(f"MAE  (unweighted): {mae_unweighted:.6f}")  # MAE
print(f"RMSE (unweighted): {rmse_unweighted:.6f}")  # RMSE
print(f"MAE  (weighted by n): {mae_weighted:.6f}")  # Í∞ÄÏ§ë MAE
print(f"RMSE (weighted by n): {rmse_weighted:.6f}")  # Í∞ÄÏ§ë RMSE
print(f"Spearman(rank corr): {spearman_corr:.6f}")  # Spearman
print(f"Top-{K} Hit count: {hit_k} / {K}")  # Hit@K
print(f"Top-{K} Jaccard  : {jaccard_k:.6f}")  # Jaccard

print("\n=== District comparison table (top 10 by predicted rate) ===")  # ÎπÑÍµê ÌÖåÏù¥Î∏î Ìó§Îçî
print(gu_cmp.sort_values("pred_rate", ascending=False).head(10)[["Íµ¨", "n", "actual_rate", "pred_rate", "abs_err"]])  # ÏÉÅÏúÑ 10Í∞ú Ï∂úÎ†•
```

```python
# ============================================================
# 10) Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞(<=2022) ÏòàÏ∏° ‚Üí Íµ¨Î≥Ñ ÏòàÏ∏° ÌèêÏóÖÎ•† ÌÖåÏù¥Î∏î ÏÉùÏÑ±(ÏßÄÎèÑ/ÎßâÎåÄÍ∑∏ÎûòÌîÑÏö©)
# ============================================================
df_pred_raw = df.copy()  # ÏòàÏ∏° Í≤∞Í≥ºÎ•º Î∂ôÏùº ÏõêÎ≥∏(Î¨∏ÏûêÏó¥ 'Íµ¨' Ìè¨Ìï®)

df_pred_enc = df_pred_raw.copy()  # ÏòàÏ∏°Ïö©(Ïù∏ÏΩîÎî© Ï†ÅÏö©) DF ÏÉùÏÑ±
for c in cat_cols:  # Î≤îÏ£ºÌòï Ïù∏ÏΩîÎî© Ï†ÅÏö©
    mapping = enc_maps[c]  # ÌïôÏäµ Îïå ÎßåÎì† mapping ÏÇ¨Ïö©
    df_pred_enc[c] = df_pred_enc[c].fillna("Unknown").astype(str)  # Í≤∞Ï∏° Ï≤òÎ¶¨
    df_pred_enc[c] = df_pred_enc[c].map(lambda x: mapping.get(x, mapping["Unknown"])).astype(int)  # Ï†ïÏàò Ïù∏ÏΩîÎî©

df_pred_raw["p_close_3y"] = clf.predict_proba(df_pred_enc[feature_cols].values)[:, 1]  # Ï†ÑÏ≤¥ ÏòàÏ∏° ÌôïÎ•†

gu_rate = df_pred_raw.groupby("Íµ¨")["p_close_3y"].mean().reset_index(name="ÏòàÏ∏°_ÌèêÏóÖÎ•†(3ÎÖÑÎÇ¥)")  # Íµ¨Î≥Ñ ÏòàÏ∏°Î•†
gu_rate["Íµ¨"] = gu_rate["Íµ¨"].astype(str).apply(lambda s: s if s.endswith("Íµ¨") else s + "Íµ¨")  # Íµ¨ ÌëúÍ∏∞ ÌÜµÏùº
```

```python
# ============================================================
# (Ï∂îÍ∞Ä) PR-AUC Ìï¥ÏÑùÏö© baseline(ÏñëÏÑ±ÎπÑÏú®) + ÎçîÎØ∏(ÏÉÅÏàòÌôïÎ•†) ÎåÄÎπÑ LogLoss/Brier Í∞úÏÑ†Ìè≠ Í≥ÑÏÇ∞
# ============================================================

import numpy as np  # ÏàòÏπò Í≥ÑÏÇ∞Ïö©
from sklearn.metrics import average_precision_score  # PR-AUC(AP)
from sklearn.metrics import log_loss  # LogLoss
from sklearn.metrics import brier_score_loss  # Brier Score

# --- Ï†ÑÏ†ú: y_test(0/1), p_test(ÏòàÏ∏°ÌôïÎ•†)Ïù¥ Ïù¥ÎØ∏ ÏûàÏùå ---
y = np.asarray(y_test).astype(int)  # y_testÎ•º intÎ°ú ÌÜµÏùº
p = np.asarray(p_test).astype(float)  # p_testÎ•º floatÎ°ú ÌÜµÏùº

# 1) ÌÖåÏä§Ìä∏ ÏñëÏÑ± ÎπÑÏú®(=ÌèêÏóÖ ÎπÑÏú®) = PR-AUCÏùò Í∏∞Ï§ÄÏÑ†(baseline)
pos_rate = y.mean()  # ÏñëÏÑ± ÎπÑÏú® Í≥ÑÏÇ∞(ÌèêÏóÖ ÎπÑÏú®)

# 2) ÌòÑÏû¨ Î™®Îç∏Ïùò PR-AUC(AP)
ap = average_precision_score(y, p)  # PR-AUC(AP) Í≥ÑÏÇ∞

# 3) ÎçîÎØ∏(ÏÉÅÏàòÌôïÎ•†) Î™®Îç∏: Î™®Îì† ÏÉòÌîåÏóê pos_rate ÌôïÎ•†ÏùÑ ÏòàÏ∏°ÌïúÎã§Í≥† Í∞ÄÏ†ï
p_dummy = np.full_like(p, fill_value=pos_rate, dtype=float)  # pos_rateÎ°ú Ï±ÑÏö¥ ÌôïÎ•† Î≤°ÌÑ∞ ÏÉùÏÑ±

# 4) ÎçîÎØ∏ ÎåÄÎπÑ LogLoss / Brier ÎπÑÍµê
ll_model = log_loss(y, p, eps=1e-15)  # Î™®Îç∏ LogLoss
ll_dummy = log_loss(y, p_dummy, eps=1e-15)  # ÎçîÎØ∏ LogLoss(Ìï≠ÏÉÅ pos_rate ÏòàÏ∏°)

brier_model = brier_score_loss(y, p)  # Î™®Îç∏ Brier
brier_dummy = brier_score_loss(y, p_dummy)  # ÎçîÎØ∏ Brier

# 5) Í∞úÏÑ†Ìè≠(ÏñëÏàòÎ©¥ Î™®Îç∏Ïù¥ Îçî Ï¢ãÏùå: ÎçîÎØ∏Î≥¥Îã§ ÏÜêÏã§Ïù¥ Ï§ÑÏóàÎã§Îäî Îúª)
ll_gain = ll_dummy - ll_model  # LogLoss Í∞úÏÑ†Ìè≠(ÎçîÎØ∏ - Î™®Îç∏)
brier_gain = brier_dummy - brier_model  # Brier Í∞úÏÑ†Ìè≠(ÎçîÎØ∏ - Î™®Îç∏)

# 6) PR-AUC Í∞úÏÑ†Ìè≠(ÏñëÏàòÎ©¥ baselineÎ≥¥Îã§ Ï¢ãÏùå)
ap_gain = ap - pos_rate  # AP - baseline

# 7) Ï∂úÎ†•
print("=== Baselines & Improvements (TEST) ===")  # Ìó§Îçî Ï∂úÎ†•
print(f"Positive rate (baseline for PR-AUC) : {pos_rate:.6f}")  # ÏñëÏÑ±ÎπÑÏú® Ï∂úÎ†•
print(f"PR-AUC(AP)                         : {ap:.6f}")  # AP Ï∂úÎ†•
print(f"AP gain (AP - positive_rate)       : {ap_gain:.6f}")  # AP Í∞úÏÑ†Ìè≠ Ï∂úÎ†•
print("")  # Ï§ÑÎ∞îÍøà
print(f"LogLoss  (model)                   : {ll_model:.6f}")  # Î™®Îç∏ LogLoss
print(f"LogLoss  (dummy=pos_rate)          : {ll_dummy:.6f}")  # ÎçîÎØ∏ LogLoss
print(f"LogLoss gain (dummy - model)       : {ll_gain:.6f}")  # LogLoss Í∞úÏÑ†Ìè≠
print("")  # Ï§ÑÎ∞îÍøà
print(f"Brier    (model)                   : {brier_model:.6f}")  # Î™®Îç∏ Brier
print(f"Brier    (dummy=pos_rate)          : {brier_dummy:.6f}")  # ÎçîÎØ∏ Brier
print(f"Brier gain (dummy - model)         : {brier_gain:.6f}")  # Brier Í∞úÏÑ†Ìè≠
```
