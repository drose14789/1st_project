```python
# ê¸ˆë¦¬ ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
base_df = pd.read_csv(
    r"C:\ai\lecNote\1st_project\data\korea_base_rate.csv",
    encoding="utf-8"
)
```

```python
# ê¸°ì¤€ê¸ˆë¦¬ ë‚ ì§œ ì •ë¦¬
base_df['date'] = pd.to_datetime(base_df['date'])
base_df = base_df.sort_values('date')

# ê¸ˆë¦¬ ì´ë²¤íŠ¸ â†’ ì—°ì›”
base_df['ì—°ì›”'] = base_df['date'].dt.to_period('M')

# ì „ì²´ ì›” ì¸ë±ìŠ¤ ìƒì„± (ë¶„ì„ ê¸°ê°„ê³¼ ë™ì¼í•˜ê²Œ)
monthly_index = pd.period_range(
    start='2000-01',
    end='2022-12',
    freq='M'
)

monthly_rate = pd.DataFrame({'ì—°ì›”': monthly_index})

# ê¸°ì¤€ê¸ˆë¦¬ ë³€ê²½ ì´ë²¤íŠ¸(base_df)ë¥¼ ì›” ë‹¨ìœ„ ê¸°ì¤€(monthly_rate)ì— ê²°í•©
# - ì—°ì›”ì„ ê¸°ì¤€ìœ¼ë¡œ ì›”ë³„ ê¸°ì¤€ê¸ˆë¦¬ë¥¼ ë§¤ì¹­
# - ë³€ê²½ì´ ë°œìƒí•œ ì›”ì—ë§Œ base_rateê°€ ì±„ì›Œì§
# - ë³€ê²½ì´ ì—†ëŠ” ì›”ì€ ì´í›„ ffill(ì´ì „ ê°’ ìœ ì§€)ë¡œ ì±„ìš¸ ì˜ˆì •
monthly_rate = monthly_rate.merge(
    base_df[['ì—°ì›”', 'base_rate']],
    on='ì—°ì›”',
    how='left'
)

# â­ í•µì‹¬: ì´ì „ ê¸ˆë¦¬ ìœ ì§€
monthly_rate['base_rate'] = monthly_rate['base_rate'].ffill()
```

```python
# df ë°ì´í„°ì—ì„œ ì¸í—ˆê°€ì¼ìì—ì„œ â€˜ì—°-ì›”â€™ë§Œ ë½‘ì•„ì„œ ìƒˆë¡œìš´ ì»¬ëŸ¼ì„ ë§Œë“œ
df['ì—°ì›”'] = df['ì¸í—ˆê°€ì¼ì'].dt.to_period('M')


# ì—°ì›”(ì›” ë‹¨ìœ„) ê¸°ì¤€ìœ¼ë¡œ ê¸°ì¤€ê¸ˆë¦¬(monthly_rate)ë¥¼ ìŒì‹ì  ë°ì´í„°(df)ì— ê²°í•©
# - dfì˜ ëª¨ë“  í–‰ì„ ìœ ì§€í•˜ê¸° ìœ„í•´ left join ì‚¬ìš©
# - ë™ì¼í•œ ì—°ì›”ì˜ base_rateê°€ dfì— ì»¬ëŸ¼ìœ¼ë¡œ ì¶”ê°€ë¨
# - í•´ë‹¹ ì›”ì— ê¸ˆë¦¬ ë°ì´í„°ê°€ ì—†ì„ ê²½ìš° base_rateëŠ” NaN
df = df.merge(
    monthly_rate,
    on='ì—°ì›”',
    how='left'
)
```

```python
KEYWORD_GROUPS = {
    'ì¹´í˜': ['ì¹´í˜', 'ì»¤í”¼', 'coffee', 'COFFEE','ì¼€ì´í¬', 'ê¹Œí˜', 'cafe', 'CAFÃ‰', 'ë‹¤ë°©', 'ìƒëŸ¬ë“œ', 'í”„ë ˆì†Œ', 'í‹°', 'ì°¨', 'ë””ì €íŠ¸'],
    'í•œì‹': ['ì‹ë‹¹', 'ì°œ', 'ì „ê³¨', 'ì „', 'ìŒì‹ì ', 'ë°¥ìƒ', 'í•œì‹', 'ìˆœëŒ€', 'ì°Œê°œ', 'êµ¬ì´' ,'ê³±ì°½', 'êµ­ìˆ˜', 'ê³ ê¸°', 'êµ¬ì´', 'êµ­', 'ë°¥', 'ë§ˆë‹¹', 'ì˜¥', 'ê´€', 'ë–¡ë³¶ì´', 'ê¹€ë°¥'],
    'ì£¼ì /ì¹˜í‚¨': ['í¬ë¦¬ìŠ¤í”¼', 'KFC', 'kfc', 'ì¹˜í‚¨', 'ë¹„ë¹„í', 'bbq',  'BBQ', 'í›„ë¼ì´ë“œ', 'ìˆ ì§‘', 'ì£¼ì ', 'í¬ì¥ë§ˆì°¨', 'ìˆ ', 'í˜¸í”„', 'ê±´ë°°', 'ë§¥ì£¼', 'í¬ì°¨','í','pup' ,'PUP', 'ì£¼ì ', 'ë°”', 'ë£¸' ,'ì™€ì¸', 'í¬ì°¨','ì†Œì£¼', 'ë§¥ì£¼', 'ì•„ì§€íŠ¸', 'ì”', 'ìœ„ìŠ¤í‚¤', 'í´ëŸ½'],
    'ì–‘ì‹/ì™¸ì‹': ['ë² ì´ì»¤ë¦¬', 'ë¹µ', 'ì•„ì›ƒë°±'],
    'ì¤‘/ì¼ì‹': ['ë§ˆë¼íƒ•','ì–‘ê¼¬ì¹˜','ìƒ¤ë¸Œ']
}
```

```python
def reclassify_etc(row):
    if row['ì—…íƒœ_ì •ë¦¬'] != 'ê¸°íƒ€':
        return row['ì—…íƒœ_ì •ë¦¬']

    text = f"{row['ì‚¬ì—…ì¥ëª…']} {row['ì—…íƒœ']}".lower()

    for group, keywords in KEYWORD_GROUPS.items():
        for kw in keywords:
            if kw.lower() in text:
                return group

    return 'ê¸°íƒ€'

df['ì—…íƒœ_ìµœì¢…'] = df.apply(reclassify_etc, axis=1)

df['ì—…íƒœ_ìµœì¢…'].value_counts()
```

```python
def map_category(x):
    if pd.isna(x):
        return 'ê¸°íƒ€'

    # ğŸ”¹ ì´ë¯¸ ëŒ€ë¶„ë¥˜ë©´ ê·¸ëŒ€ë¡œ ìœ ì§€
    if x in ['í•œì‹', 'ì¹´í˜', 'ì£¼ì /ì¹˜í‚¨', 'ì–‘ì‹/ì™¸ì‹', 'ì¤‘/ì¼ì‹']:
        return x

    # ğŸ”¹ ì„¸ë¶€ ì—…íƒœ â†’ ëŒ€ë¶„ë¥˜
    if x in ['íƒ•ë¥˜(ë³´ì‹ ìš©)', 'ëƒ‰ë©´ì§‘', 'ì‹ìœ¡(ìˆ¯ë¶ˆêµ¬ì´)']:
        return 'í•œì‹'

    if x in ['ë¶„ì‹', 'ê¹€ë°¥(ë„ì‹œë½)', 'íŒ¨ìŠ¤íŠ¸í‘¸ë“œ']:
        return 'ë¶„ì‹/ê°„í¸ì‹'

    if x in ['ê²½ì–‘ì‹', 'íŒ¨ë°€ë¦¬ë ˆìŠ¤íŠ¸ë‘', 'ì™¸êµ­ìŒì‹ì „ë¬¸ì (ì¸ë„,íƒœêµ­ë“±)', 'ë·”í˜ì‹']:
        return 'ì–‘ì‹/ì™¸ì‹'

    if x in ['ì¤‘êµ­ì‹', 'ì¼ì‹']:
        return 'ì¤‘/ì¼ì‹'

    if x in ['í˜¸í”„/í†µë‹­', 'ì •ì¢…/ëŒ€í¬ì§‘/ì†Œì£¼ë°©', 'í†µë‹­(ì¹˜í‚¨)', 'ê°ì„±ì£¼ì ', 'ê°„ì´ì£¼ì ', 'ë£¸ì‚´ë¡±']:
        return 'ì£¼ì /ì¹˜í‚¨'

    if x in ['ê¹Œí˜', 'ì»¤í”¼ìˆ', 'ì „í†µì°»ì§‘', 'ë¼ì´ë¸Œì¹´í˜', 'ë‹¤ë°©', 'í‚¤ì¦ˆì¹´í˜']:
        return 'ì¹´í˜'

    return 'ê¸°íƒ€'
```

```python
df['ì—…íƒœ_ê·¸ë£¹'] = df['ì—…íƒœ_ìµœì¢…'].apply(map_category)
```

```python
# ===============================
# ğŸ“Œ ê¸°ì¤€ ì»¬ëŸ¼ëª…
# ===============================
open_col  = "ì¸í—ˆê°€ì¼ì"
close_col = "íì—…ì¼ì"

# ===============================
# ğŸ“Œ ê´€ì¸¡ ì¢…ë£Œì¼ (ë°ì´í„° ê¸°ì¤€)
# ===============================
CUTOFF = pd.Timestamp("2022-11-30")

# ===============================
# 1ï¸âƒ£ ë‚ ì§œ íŒŒì‹±
# ===============================
df[open_col]  = pd.to_datetime(df[open_col], errors="coerce")
df[close_col] = pd.to_datetime(df[close_col], errors="coerce")

# ===============================
# 2ï¸âƒ£ ë¶„ì„ ê¸°ê°„ ì œí•œ
# ===============================
df = df[df[open_col].between("2000-01-01", CUTOFF)].copy()

# ===============================
# 3ï¸âƒ£ 3ë…„ íŒì • ê°€ëŠ¥í•œ ì°½ì—…ë§Œ ìœ ì§€ (â­ í•µì‹¬)
# ===============================
eligible_open = CUTOFF - pd.DateOffset(years=3)
df = df[df[open_col] <= eligible_open].copy()

# ===============================
# 4ï¸âƒ£ ì¢…ì†ë³€ìˆ˜ ìƒì„± (0/1)
# ===============================
open_plus_3y = df[open_col] + pd.DateOffset(years=3)

df["íì—…_3ë…„ì´ë‚´"] = (
    df[close_col].notna() &
    (df[close_col] <= open_plus_3y)
).astype("int8")

# ===============================
# 5ï¸âƒ£ ê²€ì¦
# ===============================
print("ë¼ë²¨ ë¶„í¬")
print(df["íì—…_3ë…„ì´ë‚´"].value_counts())
print("ì¸í—ˆê°€ì¼ì ë²”ìœ„:", df[open_col].min(), "~", df[open_col].max())
print("3ë…„ íŒì • ê¸°ì¤€ì¼ <=", eligible_open)
```

```python
# ===============================
# ğŸ“Œ ì—°-ì›” í‚¤ ìƒì„±
# ===============================
df["ym"] = df[open_col].dt.to_period("M").astype(str)

# ===============================
# ğŸ“Œ ì›”ë³„ íì—…ë¥ 
# ===============================
monthly = (
    df.groupby("ym")["íì—…_3ë…„ì´ë‚´"]
      .agg(
          n="size",
          close_rate="mean"
      )
      .reset_index()
)

monthly["close_rate_pct"] = (monthly["close_rate"] * 100).round(2)

print(monthly.head())
print("shape:", monthly.shape)

# ===============================
# ğŸ“Œ ì¢…ì†ë³€ìˆ˜ y
# ===============================
y = monthly["close_rate"].values
print("y shape:", y.shape)
```

```python
import pandas as pd
import numpy as np
import re

rate_path = r"C:\ai\lecNote\1st_Project\data\ê¸ˆë¦¬,ì‹œì¥ë©´ì ì¶”ê°€.csv"
emp_path  = r"C:\ai\lecNote\1st_Project\data\ê³ ìš©ì§€í‘œ_ìˆ˜ì •.csv"

rate = pd.read_csv(rate_path, low_memory=False)
emp  = pd.read_csv(emp_path, low_memory=False)

rate_dt = pd.to_datetime(rate["ì¸í—ˆê°€ì¼ì"], errors="coerce")
rate["ì¸í—ˆê°€ì—°ì›”"] = (rate_dt.dt.year * 100 + rate_dt.dt.month).astype("Int64")

def parse_ym(s):
    if pd.isna(s): return np.nan
    s = str(s).strip()
    m = re.search(r"(\d{4})\D*(\d{1,2})", s)  # '2000. 01'
    if not m: return np.nan
    y, mo = int(m.group(1)), int(m.group(2))
    return y * 100 + mo

emp["ì¸í—ˆê°€ì—°ì›”"] = emp["ì‹œì "].apply(parse_ym).astype("Int64")

col_map = {
    "ê²½ì œí™œë™ì°¸ê°€ìœ¨": "ì¸í—ˆê°€ì¼ì_ê²½ì œí™œë™ì°¸ê°€ìœ¨",
    "ì‹¤ì—…ë¥ ": "ì¸í—ˆê°€ì¼ì_ì‹¤ì—…ë¥ ",
    "ê³ ìš©ë¥ ": "ì¸í—ˆê°€ì¼ì_ê³ ìš©ë¥ ",
}
emp2 = emp.rename(columns=col_map)[["ì¸í—ˆê°€ì—°ì›”"] + list(col_map.values())]
emp2 = emp2.dropna(subset=["ì¸í—ˆê°€ì—°ì›”"]).drop_duplicates("ì¸í—ˆê°€ì—°ì›”", keep="last")

merged = rate.merge(emp2, on="ì¸í—ˆê°€ì—°ì›”", how="left")

merged.to_csv(r"C:\ai\lecNote\1st_Project\data\ê¸ˆë¦¬ì‹œì¥ë©´ì _ê³ ìš©ì§€í‘œìˆ˜ì •_ì¸í—ˆê°€ì¼ìë¨¸ì§€.csv",
              index=False, encoding="utf-8-sig")
```
