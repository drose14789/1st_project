```python
# ============================================================
# 2) 데이터 로드 + 라벨 정리(0/1만 사용, 3은 제외)
# ============================================================
DATA_PATH = r"C:\ai\lecNote\1st_Project\data\서울_일반음식점_라벨3처리.csv"  # 데이터 파일 경로 지정
df_raw = pd.read_csv(DATA_PATH, low_memory=False)  # CSV 로드

if "폐업_3년이내" not in df_raw.columns:  # 타깃 컬럼 존재 확인
    raise ValueError("컬럼 '폐업_3년이내'가 없습니다.")  # 없으면 즉시 중단

df = df_raw.copy()  # 원본 보존을 위해 복사
df = df[df["폐업_3년이내"].isin([0, 1])].copy()  # 라벨 0/1만 사용(3 제외)
df["y"] = df["폐업_3년이내"].astype(int)  # 학습용 타깃 생성
```
